# Import Python Libraries

```python
# Import all packages in this section
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read Data


```python
# Reading HR attrition Dataset
df = pd.read_excel("/content/HRAttrition.xlsx", sheet_name='Employee Data 2018 - 2019')
```

# Data Exploration - Display the basic information

```python
df.info()
```

# Data Exploration - Descriptive Statistics

```python
df.describe().transpose()
```

```python
df["Attrition_int"] = df["Attrition"].replace({"Yes": 1, "No": 0})
df_attrtion = df.groupby('Department')['Attrition_int'].sum().reset_index(name='# Attrition')
df_attrtion.sort_values(by='# Attrition', ascending=False, inplace=True)
df_attrtion
```

# Plot the graph

```python
plt.figure(figsize=(10, 6))
df_attrtion.plot(kind='bar', x='Department', color='skyblue')

plt.title('Attrition Count by Department')
plt.xlabel('Department')
plt.ylabel('Attrition Count (#)')
plt.xticks(rotation=360)
plt.tight_layout()
plt.show()
```
